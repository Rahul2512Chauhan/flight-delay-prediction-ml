##3.2 . Predictive Model:

###Data Preparation

####Create Classification Target (is_delayed)

```python
# 1 if atleast one flight delayed(15+ min) , else 0
df['is_delayed'] = df['arr_del15'].apply(lambda x: 1 if x>=1 else 0)
```

####Handle Missing Values (Numerical Columns Only)

```python
# Select numeric columns only
num_cols = df.select_dtypes(include='number').columns

# Impute missing values with median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

####Feature Selection

```python
# Target column
y = df['is_delayed']

# Drop non-numeric + leakage columns
columns_to_drop = [
    'carrier_name', 'airport_name',
    'arr_del15', 'arr_delay', 'total_delay', 'is_delayed',
    'carrier_ct', 'weather_ct', 'nas_ct', 'security_ct', 'late_aircraft_ct',
    'arr_cancelled', 'arr_diverted',
    'carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay',
    'delay_rate'
]

# Drop these from df to form X
X = df.drop(columns=columns_to_drop)

# Select only numeric columns from the cleaned DataFrame X
X = X.select_dtypes(include='number')
```

####Feature Scaling (optional for tree models, but useful if using logistic regression or others)

```python
print("Original feature columns:", X.columns.tolist())
print("Shape before scaling:", X.shape)
```

####Train-Test Split (80/20)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_array = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled_array, columns=X.columns)
print("Shape after scaling (X_scaled):", X_scaled.shape)
```

# Classification Modeling

### Train a Random Forest Classifier

```python
X_df = pd.DataFrame(X_scaled, columns=X.columns)  # restore DataFrame with names
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_df, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and fit model
clf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
clf.fit(X_train, y_train)
```

### Model Evaluation

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_auc_score, RocCurveDisplay
)
import matplotlib.pyplot as plt
import seaborn as sns

# Predict
y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:, 1]

# Metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# ROC curve
RocCurveDisplay.from_estimator(clf, X_test, y_test)
plt.title("ROC Curve")
plt.show()
```

### Operatinal Adjustability Index

#### Assign Adjustability Scores

```python
# Define adjustability scores (custom, based on domain knowledge)
adjustability_scores = {
    'carrier_ct': 1.0,           # Fully controllable (airline ops)
    'weather_ct': 0.0,           # Not controllable
    'nas_ct': 0.3,               # Partially controllable (airspace, ATC delays)
    'security_ct': 0.4,          # Somewhat controllable (airport ops)
    'late_aircraft_ct': 0.6,     # Upstream control possible
    'arr_flights': 0.5,          # Scheduling / capacity planning
    'month': 0.0,                # Not adjustable
    'year': 0.0                  # Not adjustable
}
```

#### Extract Feature Importances from Random Forest

```python
import pandas as pd

# Assuming you already have the trained RandomForestClassifier as `clf`
# And your input dataframe before scaling as `X_df` (with proper feature names)

feature_names = X_df.columns
importances = clf.feature_importances_

# Create a DataFrame to hold feature names and their importance
feat_imp_df = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
})
```

#### Compute OAI = Importance × Adjustability

```python
# Map adjustability scores
feat_imp_df['adjustability'] = feat_imp_df['feature'].map(adjustability_scores).fillna(0)

# Compute Operational Adjustability Index
feat_imp_df['OAI'] = feat_imp_df['importance'] * feat_imp_df['adjustability']
```

#### Sort and View Results

```python
# Sort features by OAI
feat_imp_df_sorted = feat_imp_df.sort_values(by='OAI', ascending=False)

# Show final ranked features
print(feat_imp_df_sorted[['feature', 'importance', 'adjustability', 'OAI']])
```

Insights:

arr_flights is your top feature:

High importance (86% of model decision weight).

Adjustability = 0.5 → You have some operational control over it (scheduling, fleet management, staffing).

High OAI = 0.432 → Most impactful and controllable → ✅ Prioritize this for intervention.

month and year:

Somewhat important to the model, but OAI = 0 because you can't adjust them.

This is expected — these are seasonal/contextual features, not operational levers.

# Regression Model (XGBoost Regression on Total Delay)

### Setup

```python
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np
```

### Prepare Features and Target

```python
X = df.drop(columns=[
    'carrier_name', 'airport_name', 'arr_del15', 'arr_delay', 'is_delayed',
    'carrier_ct', 'weather_ct', 'nas_ct', 'security_ct', 'late_aircraft_ct',
    'arr_cancelled', 'arr_diverted',
    'carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay',
    'delay_rate', 'total_delay'   
])

X = X.select_dtypes(include='number')
y_reg = df['total_delay'].clip(upper=500)
```

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)
```

### Train XGBoost Regressor

```python
model = xgb.XGBRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train, y_train)

```

### Evaluate Performance

```python
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MAE : {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
residuals = y_test - y_pred

plt.figure(figsize=(8, 4))
sns.histplot(residuals, kde=True, bins=30, color='tomato')
plt.title("Residual Distribution (Actual - Predicted)")
plt.xlabel("Residual (minutes)")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.5, edgecolor=None)
plt.plot([0, max(y_test)], [0, max(y_test)], color='black', linestyle='--')
plt.xlabel("Actual Delay (minutes)")
plt.ylabel("Predicted Delay (minutes)")
plt.title("Actual vs Predicted Delay")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Operational Adjustability Index (OAI) Computation

```python
import shap
import pandas as pd

# Create SHAP explainer and get values
explainer = shap.Explainer(model, X_train)
X_sample = X_train.sample(100, random_state=42)
shap_values = explainer(X_sample)

# Mean absolute SHAP values across samples
shap_df = pd.DataFrame({
    'feature': X_sample.columns,
    'shap_value': np.abs(shap_values.values).mean(axis=0)
})

```

```python
# Assign weights
adjustability_weights = {
    'arr_flights': 0.5,
    'year': 0.0,
    'month': 0.0
}

shap_df['adjustability'] = shap_df['feature'].map(adjustability_weights).fillna(0)
```

```python
shap_df['OAI_weighted'] = shap_df['shap_value'] * shap_df['adjustability']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

shap_df_sorted = shap_df.sort_values(by='OAI_weighted', ascending=False)

plt.figure(figsize=(6, 4))
sns.barplot(
    x='OAI_weighted',
    y='feature',
    data=shap_df_sorted,
    hue='adjustability',
    palette='viridis'
)
plt.title('OAI-Weighted SHAP Feature Importance')
plt.tight_layout()
plt.show()
```
